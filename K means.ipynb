```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=11)
plt.figure(0)
plt.grid(True)
plt.scatter((X[:,0]),X[:,1])
plt.show()
```

```python
k = 5
colors = ["green","red","blue","yellow","orange"]

clusters = {}

for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    
    
    cluster = {
        'center':center,
        'points':points,
        'color':colors[i]
        }
    
    clusters[i] = cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointstoClusters(clusters):
    for ix in range(X.shape[0]):
        dist = []
        curr_X = X[ix]

        for kx in range(k):
            d = distance(curr_X,clusters[kx]['center'])
            dist.append(d)

        current_cluster = np.argmin(dist)
        clusters[current_cluster]['points'].append(curr_X)
```

```python
def updateClusters(clusters):
     for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if(pts.shape[0]>0):
            new_u = pts.mean(axis=0)
            clusters[kx]['center'] = new_u
            clusters[kx]['points'] = []   
```

```python
def plotClusters(clusters):
    
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        uk = clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color='black',marker='*')
```

```python
plotClusters(clusters)
```

```python
assignPointstoClusters(clusters)
plotClusters(clusters)
updateClusters(clusters)
```

## Using Sklearn

```python
from sklearn.cluster import KMeans
```

```python
X,Y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=9)
```

```python
kmeans= KMeans(n_clusters=5)
```

```python
kmeans.fit(X,Y)
centers = kmeans.cluster_centers_
```

```python
pred = kmeans.labels_
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.show()
```
